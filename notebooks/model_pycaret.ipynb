```python
import pandas as pd
from pycaret.classification import *

df = pd.read_csv('../data/clean_data.csv')

clf = setup(
    data=df,
    target='label',
    text_features=['text'],
    session_id=42
)
```

```python
lr = create_model('lr')
nb = create_model('nb')
svm = create_model('svm')
```

```python
compare_models(include=['lr', 'nb', 'svm'])
```

```python
best_model = create_model('svm')
tuned_model = tune_model(best_model)
final_model = finalize_model(tuned_model)
```

```python
save_model(final_model, '../models/sentiment_model')
save_model(final_model, '../sentiment_model')
```
